```python
#sum of digits
num = int(input("Enter the number: "))
sum = 0
while num > 0:
  sum = sum + (num % 10)
  num = num // 10
print("Sum of digits is : ", sum)
```

```python
#product of digits
num = int(input("Enter the number: "))
pro = 1
while num > 0:
  pro = pro * (num % 10)
  num = num // 10
print("Product of digits is :", pro)
```

```python
#reverse a number
num = int(input("Enter the number: "))
rev = 0
num1 = num
while num > 0:
  rem = num % 10
  rev = rev * 10 + rem
  num = num // 10
print("Reverse of", num1, "is : ", rev)
```

```python
#palindrome of a number
num = int(input("Enter the number:"))
rev = 0
num1 = num
while num > 0:
  rem = num % 10
  rev = rev * 10 + rem
  num = num // 10
if rev == num1:
  print(rev, "is a palindrome of", num1)
else:
  print(rev, "is not a palindrome of", num1)
```

```python
#armstrong number
num = int(input("Enter the number:"))
res = 0
num1 = num
power = len(str(num))
while num > 0:
  rem = num % 10
  res = res + rem ** power
  num = num // 10
if res == num1:
  print(num1, "is an armstrong number")
else:
  print(num1, "is not an armstrong number")
```
